```python
import pandas as pd

df = pd.read_csv("/content/Match_Result_Data 2017-24.csv")
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(['notes', 'attendance', 'formation'], axis=1)
```

```python
df1.columns
```

Dataframe for the year 2020

```python
df_2020 = df1[df1['date'].str.contains('2020')]
df_2020
```

Dataframe for the year 2021

```python
df_2021 = df1[df1['date'].str.contains('2021')]
df_2021
```

Dataframe for the year 2022

```python
df_2022 = df1[df1['date'].str.contains('2022')]
df_2022
```

```python
import matplotlib.pyplot as plt

# Group by team and count matches played, wins, and losses in 2020
team_stats_2020 = df_2020.groupby('team')['result'].agg(['count', lambda x: (x == 'W').sum(), lambda x: (x == 'L').sum()])
team_stats_2020.columns = ['Matches Played', 'Wins', 'Losses']

# Create a bar plot
team_stats_2020[['Wins', 'Losses']].plot(kind='bar', figsize=(10, 6))
plt.title('Team Matches Played, Wins, and Losses in 2020')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataframe is named 'df' and it contains 'team', 'result', 'venue' columns
# and the data for 2020 is filtered into df_2020

# Create a new column to identify home or away matches
df_2020['Home/Away'] = 'Away'
df_2020.loc[df_2020['venue'].str.contains('Home'), 'Home/Away'] = 'Home'

# Pivot table to get the counts of wins for each team at home and away
team_wins_2020 = pd.pivot_table(df_2020[df_2020['result'] == 'W'], values='result', index='team', columns='Home/Away', aggfunc='count')
team_wins_2020 = team_wins_2020.fillna(0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(team_wins_2020, annot=True, cmap='viridis', fmt='.0f')
plt.title('Team Home and Away Wins in 2020')
plt.xlabel('Home/Away')
plt.ylabel('Team')
plt.show()
```

```python
# Group by team and count matches played and losses
team_stats = df_2020.groupby('team')['result'].agg(['count', lambda x: (x == 'L').sum()])
team_stats.columns = ['Matches Played', 'Losses']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(team_stats['Matches Played'], team_stats['Losses'])
plt.title('Team Matches Played vs. Losses in 2020')
plt.xlabel('Matches Played')
plt.ylabel('Losses')
for team, row in team_stats.iterrows():
    plt.annotate(team, (row['Matches Played'], row['Losses']))
plt.show()
```

```python
# Group by team and count matches played and wins
team_stats = df_2020.groupby('team')['result'].agg(['count', lambda x: (x == 'W').sum()])
team_stats.columns = ['Matches Played', 'Wins']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(team_stats['Matches Played'], team_stats['Wins'])
plt.title('Team Matches Played vs. Wins in 2020')
plt.xlabel('Matches Played')
plt.ylabel('Wins')
for team, row in team_stats.iterrows():
    plt.annotate(team, (row['Matches Played'], row['Wins']))
plt.show()
```

```python
# Filter the dataframe for Arsenal matches in 2020
arsenal_2020 = df_2020[df_2020['team'] == 'Arsenal']

# Create a new column to identify home or away matches
arsenal_2020['Home/Away'] = 'Away'
arsenal_2020.loc[arsenal_2020['venue'].str.contains('Home'), 'Home/Away'] = 'Home'

# Calculate the number of home and away wins
home_wins = len(arsenal_2020[(arsenal_2020['result'] == 'W') & (arsenal_2020['Home/Away'] == 'Home')])
away_wins = len(arsenal_2020[(arsenal_2020['result'] == 'W') & (arsenal_2020['Home/Away'] == 'Away')])

# Create the pie chart
labels = ['Home Wins', 'Away Wins']
sizes = [home_wins, away_wins]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # Explode the 1st slice (Home Wins)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Arsenal Home and Away Wins in 2020')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
